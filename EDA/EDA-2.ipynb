```python
print(df['subject_id'].nunique())
print(df['night_id'].nunique())
print(df.groupby('night_id')['epoch_index'].nunique())
```

```python
numeric_cols = ['age', 'bmi', 'acc_mean', 'acc_std', 'hr_mean', 'hr_std', 'spo2_mean', 'spo2_min']
print("Summary Statistics for Physiological Features:")
display(df[numeric_cols].describe()) 
```

```python
event_counts = df['is_apnea_event'].value_counts()
event_percentages = df['is_apnea_event'].value_counts(normalize=True) * 100
print("\nDistribution of 'is_apnea_event':")
print(f"  0 (No Apnea):   {event_counts[0]:,} rows ({event_percentages[0]:.1f}%)")
print(f"  1 (Apnea Event): {event_counts[1]:,} rows ({event_percentages[1]:.1f}%)")
```

```python
severity_counts = df['severity_label'].value_counts()
print("Distribution of 'severity_label':")
for severity, count in severity_counts.items():
    print(f"  {severity}: {count:,} rows ({count/len(df)*100:.1f}%)")
```

```python
print("Cross-tab: severity_label vs is_apnea_event:")
cross_tab = pd.crosstab(df['severity_label'], df['is_apnea_event'], normalize='index') * 100
print(cross_tab.round(1))
```

```python
print("Subject-wise Severity Distribution:")
subject_severity = df.groupby(['subject_id', 'severity_label']).size().unstack(fill_value=0)
print(subject_severity)
```
